```python
%%writefile satnews.py
import scrapy


class SatnewsSpider(scrapy.Spider):
    name = "satnews"
    allowed_domains = ["www.tunisia-sat.com"]
    start_urls = ["https://www.tunisia-sat.com/forums/news/"]
    def parse(self, response):
      news=response.css('div.structItem-cell')
      for new in news:
        yield{
                'title' : new.css('div.structItem-title a::text').get(),
                'categorie' :new.css('div.structItem-title span::text').get(),
                'time' :new.css('li.structItem-startDate a time.u-dt::text').get(),
                'publisher' : new.css('ul.structItem-parts a::text').get(),
                }
```
