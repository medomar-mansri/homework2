```python
%%writefile yahoonews.py
import scrapy
class YahoonewsSpider(scrapy.Spider):
    name = "yahoonews"
    allowed_domains = ["www.yahoo.com"]
    start_urls = ["https://www.yahoo.com/news/"]

    def parse(self, response):
      news=response.css('li.stream-item')
      for new in news:
        yield{
                'title' : new.css('h3 a::text').get(),
                'url' : new.css('h3 a').attrib['href'],
                'description' :new.css('p::text').get(),
              }
```
